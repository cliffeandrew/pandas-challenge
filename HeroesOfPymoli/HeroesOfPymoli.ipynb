```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_number_of_players = len(purchase_data["SN"].unique())
player_count_df = pd.DataFrame({"Total Players":[total_number_of_players]})
player_count_df                    
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
number_of_unique_items = len(purchase_data["Item ID"].unique())
average_purchase_price = purchase_data["Price"].mean()
count_purchase_id = len(purchase_data["Purchase ID"])
total_revenue = purchase_data["Price"].sum()

purchasing_analysis_df = pd.DataFrame({"Number of Unique Items":[number_of_unique_items],
                                       "Average Price":[average_purchase_price],
                                       "Number of Purchases":[count_purchase_id],
                                       "Total Revenue":[total_revenue]})

purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:,.2f}".format)

purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_df = purchase_data.groupby(["Gender"])

gender_count = gender_df["SN"].nunique()
total_number_of_players = len(purchase_data["SN"].unique())
gender_percent = gender_count/total_number_of_players*100

gender_summary_df = pd.DataFrame({"Gender Percent": gender_percent, "Total Count":gender_count})

gender_summary_df = gender_summary_df[["Total Count", "Gender Percent"]]

gender_summary_df["Gender Percent"] = gender_summary_df["Gender Percent"].map("{:.2f}%".format)

gender_summary_df = gender_summary_df.sort_values("Gender Percent", ascending=False)

gender_summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchasing_analysis_df = purchase_data.groupby(["Gender"])

purchase_id_count = purchasing_analysis_df["Purchase ID"].nunique()
average_purchase_price = purchasing_analysis_df["Price"].mean()
total_revenue = purchasing_analysis_df["Price"].sum()

average_purchase_price = total_revenue/purchase_id_count

gender_count = purchasing_analysis_df["SN"].nunique()
avg_total_purchase_per_person = total_revenue/gender_count

purchasing_analysis_df = pd.DataFrame({"Purchase Count": purchase_id_count, "Average Purchase Price": average_purchase_price, "Total Purchase Value":total_revenue, "Avg Total Purchase per Person":avg_total_purchase_per_person})

purchasing_analysis_df = purchasing_analysis_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]

purchasing_analysis_df["Average Purchase Price"] = purchasing_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_df["Total Purchase Value"] = purchasing_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_df["Avg Total Purchase per Person"] = purchasing_analysis_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

purchasing_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head()
```

```python
top_spender_df = purchase_data.groupby(["SN"])

sn_count = top_spender_df["SN"].nunique()
purchase_id_count = top_spender_df["Purchase ID"].nunique()
total_purchase_value = top_spender_df["Price"].sum()
average_purchase_price = total_purchase_value/purchase_id_count

top_spender_summary_df = pd.DataFrame({"SN": sn_count,"Purchase Count": purchase_id_count, "Average Purchase Price": average_purchase_price,"Total Purchase Value":total_purchase_value})

top_spender_summary_df["Average Purchase Price"] = top_spender_summary_df["Average Purchase Price"].map("${:,.2f}".format)
top_spender_summary_df["Total Purchase Value"] = top_spender_summary_df["Total Purchase Value"].map("${:,.2f}".format)

top_spender_summary_df = top_spender_summary_df.sort_values("Total Purchase Value", ascending=False)
#top_spender_summary_df = top_spender_summary_df.set_index("SN")

top_spender_summary_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head()
```

```python
popular_df = purchase_data.groupby(["Item ID"])
#popular_df = purchase_data.groupby.loc([:,4:5])

item_id = popular_df["Item ID"].nunique()
item_name = popular_df["Item Name"].nunique()
purchase_count = popular_df["Purchase ID"].count()
item_price = popular_df["Price"].nunique()
total_purchase_value = popular_df["Price"].sum()

popular_summary_df = pd.DataFrame({"Item ID": item_id,"Item Name": item_name, "Purchase Count": purchase_count,"Item Price": item_price, "Total Purchase Value":total_purchase_value})

popular_summary_df["Item Price"] = popular_summary_df["Item Price"].map("${:,.2f}".format)
popular_summary_df["Total Purchase Value"] = popular_summary_df["Total Purchase Value"].map("${:,.2f}".format)

popular_summary_df = popular_summary_df.sort_values("Total Purchase Value", ascending=False)
popular_summary_df.head(5)
```
